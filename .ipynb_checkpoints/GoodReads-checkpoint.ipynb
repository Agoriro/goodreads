## Import necessary modules


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Read file

```python
df = pd.read_csv('./books.csv', error_bad_lines = False)
df
```

```python
df.info()
```

```python
df.describe(include='all')
```

### Mostrar que libro obtuvo el mayor número de reviews

```python
data = df.groupby('title')['text_reviews_count'].sum().sort_values(ascending=False).head(5)
print(data)
```

```python
#fig, ax = plt.subplots()
#ax.barh(data.index, data)

the_most_reviews = df.sort_values('text_reviews_count', ascending=False).head(5)
plt.figure(figsize=(15,10))
sns.barplot(the_most_reviews['title'], the_most_reviews['text_reviews_count'], alpha=.9,palette='rocket')
plt.show()
```

## Mostrar que libro tiene mejor puntuacion

```python
data1 = df.groupby('title')['average_rating'].sum().sort_values(ascending=False).head(5)
print(data1)
```

```python
#fig, ax1 = plt.subplots()
#ax1.barh(data1.index, data1)

the_most_average = df.sort_values('average_rating', ascending=False).head(5)
plt.figure(figsize=(15,10))
sns.barplot(the_most_average['average_rating'], the_most_average['title'], alpha=.9,palette='rocket')
plt.show()
```

## Mostrar que libro tiene mayor número de reviews con mejor puntuación

```python
data3 = df.loc[:,['title','text_reviews_count','average_rating']].sort_values(['text_reviews_count','average_rating'],ascending=False).head(5)
data3
```
